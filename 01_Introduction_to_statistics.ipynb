```
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```
rv_norm = stats.norm(loc=0, scale = 1)
```

```
#Make a sample which consists of 30 obervations
samples = rv_norm.rvs(size = 30)
```

```
print("Sample Mean : {0}, {0}".format(np.mean(samples),2))
print("Sample SD : {}".format(np.std(samples)))
```

```
#plot a PDF
xx = np.linspace(-5,5,100) # (최소, 최대, 점의 개수)
plt.plot(xx,rv_norm.pdf(xx)) 
plt.axvline(x=0, color = 'r', linestyle = '--', linewidth=1)
plt.title("PDF of normal distribution")
plt.show()
```

```
#plot a CDF
xx = np.linspace(-5,5,100) # (최소, 최대, 점의 개수)
plt.plot(xx,rv_norm.cdf(xx)) 
plt.axvline(x=0, color = 'r', linestyle = '--', linewidth=1)
plt.title("PDF of normal distribution")
plt.show()
```
